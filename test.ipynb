```python
import requests
from PIL import Image
from transformers import BlipProcessor, BlipForConditionalGeneration

processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-large")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-large")
```

```python
import matplotlib.pyplot as plt

images_to_process = [
    "/home/ajeet/codework/dataset_frames/2572904/0_2783.jpg",
    "/home/ajeet/codework/dataset_frames/2572904/0_3022.jpg",
    "/home/ajeet/codework/dataset_frames/2572904/0_3026.jpg",
    # "/home/ajeet/codework/dataset_frames/2572904/0_1100.jpg",
    # "/home/ajeet/codework/dataset_frames/2572904/0_1185.jpg",

    # "/home/ajeet/codework/dataset_frames/2572904/0_33.jpg",
    # "/home/ajeet/codework/dataset_frames/2572904/0_1182.jpg",

    # "/home/ajeet/codework/dataset_frames/2572904/0_1532.jpg",
    # "/home/ajeet/codework/dataset_frames/2572904/0_2316.jpg",
    # "/home/ajeet/codework/dataset_frames/2572904/0_2296.jpg",
    # "/home/ajeet/codework/dataset_frames/2572904/0_28.jpg",
    # "/home/ajeet/codework/dataset_frames/2572904/0_357.jpg",
    # "/home/ajeet/codework/dataset_frames/2572904/0_23.jpg",
    # "/home/ajeet/codework/dataset_frames/2572904/0_2800.jpg",
    # "/home/ajeet/codework/dataset_frames/2572904/0_2075.jpg",



]

for image_path in images_to_process:

    raw_image = Image.open(image_path).convert('RGB')

    # conditional image captioning
    text = "A photo of"
    inputs = processor(raw_image, text, return_tensors="pt")

    out = model.generate(**inputs)
    print(processor.decode(out[0], skip_special_tokens=True))

    # text = "An image showing"
    # inputs = processor(raw_image, text, return_tensors="pt")

    # out = model.generate(**inputs)
    # print(processor.decode(out[0], skip_special_tokens=True))

    # unconditional image captioning
    inputs = processor(raw_image, return_tensors="pt")

    out = model.generate(**inputs)
    print(processor.decode(out[0], skip_special_tokens=True))


    small_image = raw_image.resize((64, 64))

    # Display the image at 64x64 pixels
    plt.figure(figsize=(0.64, 0.64))  # Set figure size for 64x64 pixels
    plt.imshow(small_image)
    plt.axis('off')  # Hide axes
    plt.show()  # Show the image

    print("--------------------------------")
```
